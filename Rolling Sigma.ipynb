```python
import pandas as pd
import numpy as np

def limit_sigma(sigma, stdev_min, stdev_max):
    if sigma < stdev_min:
        return stdev_min
    elif sigma > stdev_max:
        return stdev_max
    else:
        return sigma

data = pd.DataFrame(np.array([5232,5232,5265,5265,4521,5321,5266,5273,5271,5250,5264,5282,5291,5266,5290,5287,5282,5301,5278,5278,5298,5274,5292,5295,5290,5241,5292,5296,5263,5279,5253,5268,3807,5447,5277,5291,5277,5277,5288,5281,5301,5296,5284,5294,5301,5302,5285,5289,5289,5285]),
                             columns=['LEFT'])

data['SIGMA_LEFT'] = data['LEFT'].shift().rolling(window=15, min_periods=1).std()

q1, q3 = data['SIGMA_LEFT'].quantile([0.25, 0.75])
data['SIGMA_LEFT'] = data['SIGMA_LEFT'].apply(limit_sigma, args=(q1, q3))

display(q1,q3)
display(data)
```
